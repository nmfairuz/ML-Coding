# Principle Component Analysis

Suppose there are 100 features in a dataset. We can't visualize 100 dimensions, and some features may contain simiar infos.

Then PCA asks:

> Can I represent this data using fewer features while keeping most of the important information?

>PCA can combine co-related features into one new feature that captures most of the variation.

Think of your data as a cloud of points.

PCA looks for the direction where the data spreads the most.

That direction is called Principal Component 1 (PC1).

Then it finds another direction, perpendicular to PC1, with the next largest spread.

That's PC2.

```
     •
   •
 •
•
     •
       •
```
The points clearly stretch diagonally.

PCA rotates the coordinate system so that:
```
PC2
↑
|
|
|
+------------→ PC1
```

Instead of measuring with the original x/y axes, it measures along the direction where the data naturally varies most.



````python
from sklearn.datasets import make_blobs
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
````

## Create Dataset:

````python
X, y = make_blobs(
    n_samples=400,
    centers=4,
    cluster_std=1.2,
    random_state=42
)
````

## Visualize Raw Data:

````python
plt.scatter(X[:, 0], X[:, 1])
plt.title("Original Dataset")
plt.show()
````

## Scale the data


````python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
````

## Visualize scaled data


````python
plt.scatter(X_scaled[:, 0], X_scaled[:, 1])
plt.title("Scaled Dataset")
plt.show()
````

X axis now spreads from -2 to 2 and Y axis spreads from -2 to 2

## Do PCA

````python
from sklearn.decomposition import PCA

pca=PCA(n_components=2)
X_pca=pca.fit_transform(X_scaled)
print("Explained variance ratio : ", pca.explained_variance_ratio_)
````

## Plot the result

````python
plt.scatter(X_pca[:, 0], X_pca[:, 1])
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("PCA Projection")
plt.show()
````
